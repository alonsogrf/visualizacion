Se crea una lista con las columnas de la database para utilizar dichas variables y obtener su correlación con la variable "liked"

```python
list = df.columns.tolist()
list.sort()
list
```

Se observa que hay una variable que presenta mayor correlación que las demás con la variable de interés: "speechiness"

```python
vars = ['liked','acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'key', 'liveness',
        'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence']
df = df[vars].dropna()
df[vars].corr()
```

Pero antes de visualizar su correlación, se busca codificar el ranking de acepación de la variable "liked" en 4 categorías dependiendo su clase: 1a, 2a, 3a o regular.

```python
def category(rank):
    if (rank>=0)&(rank<=0.25):
        return 'regular class'
    elif (rank>0.25)&(rank<=0.5):
        return 'third class'
    elif (rank>0.5)&(rank<=0.75):
        return 'second class'
    else:
        return 'first class'
```

```python
df['Rank'] = df['liked'].apply(lambda x:category(x))
```

Entonces ahora sí se realiza la Scatter Plot para "Speechiness" vs "Liked" categorizando en la nueva variable "Rank"

```python
sns.lmplot(x = 'speechiness', y = 'liked', data = df, fit_reg = False, hue = 'Rank')
plt.show()
```

Previo al análisis de regresión se evalúa la presencia de datos atípicos, se presentan 3 pero aún así se continúa con el modelo de predicción.

```python
sns.boxplot(data = df.loc[:, 'speechiness'])
plt.show()
```

Se observa en el modelo que los parámetros B0 y B1 son significantes, pues se encuentran por debajo del (alfa = 0.05), lo cual implica que se está manejando un buen modelo. En un análisis más detallado, el Durbin-Watson tiene un valor muy cercano a 2, lo cual también refiere a un modelo bueno, así como el valor de R cuadrada que es bajo.

```python
model = sm.OLS.from_formula('liked~speechiness', data=df)

result = model.fit()
result.summary()
```

Finalmente se grafica de nuevo la Scatter Plot con el pronóstico de regresión y se observa que se ajusta ciertamente bien para conocer el nivel de aceptación de una canción en función de la letra de la misma.

```python
lm = linear_model.LinearRegression()
X = pd.DataFrame(df['speechiness'])
Y = pd.DataFrame(df['liked'])

model_lm = lm.fit(X,Y)
model_lm

df.plot(kind='scatter',
       x = 'speechiness',
       y = 'liked',
       color = 'r')

plt.plot(X,
        model_lm.predict(X),
        linewidth = 5,
        color = 'yellow')
plt.show()
```
