```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm

mpl.style.use('ggplot')

df=pd.read_csv('datasets/antropometria-dataset.csv')
df.head(3)
```

```python
df.columns
```

```python
vars = ['diastol','sistol','edad','peso','cintura']
df = df[vars]
df.head(3)
```

```python
df = df[vars].dropna()
df.head(3)
```

```python
df[vars].corr()
```

```python
model = sm.OLS.from_formula('diastol~sistol+edad+peso+cintura', data=df)

result = model.fit()
result.summary()
```
