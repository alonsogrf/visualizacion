```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(10)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(x = df['sistol'], bins = 10)
plt.title("Sistol")

plt.subplot(1,2,2)
plt.hist(x = df['diastol'], bins = 10)
plt.title("Diastol")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(x = df['peso'], bins = 10)
plt.title("Peso")

plt.subplot(1,2,2)
plt.hist(x = df['cintura'], bins = 10)
plt.title("Cintura")

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = df['peso'], y = df['cintura'])
plt.xlabel('Peso')
plt.ylabel('Cintura')
plt.title('Diagrama de dispersión de peso vs. cintura')

plt.show()
```

```python
#Separar datos de hombres y mujeres
hombres = df.loc[df["sexo"]==1]
mujeres = df.loc[df["sexo"]==2]
```

```python
# Guardar los datos de las variables de interés
peso_hombres = hombres["peso"]
peso_mujeres = mujeres["peso"]

cintura_hombres = hombres["cintura"]
cintura_mujeres = mujeres["cintura"]
```

```python
#Mostrar 10 registros 
hombres.head(10)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(x = peso_hombres, bins = 10)
plt.title("Peso hombres")

plt.subplot(1,2,2)
plt.hist(x = peso_mujeres, bins = 10)
plt.title("Peso mujeres")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(x = cintura_hombres, bins = 10)
plt.title("Cintura hombres")

plt.subplot(1,2,2)
plt.hist(x = cintura_mujeres, bins = 10)
plt.title("Cintura mujeres")

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = peso_hombres, y = cintura_hombres)
plt.xlabel("'Peso hombres'")
plt.ylabel("'Cintura hombres'")
plt.title("Diagrama de dispersión de peso hombres vs.cintura hombres")

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = peso_mujeres, y = cintura_mujeres)
plt.xlabel("'Peso mujeres'")
plt.ylabel("'Cintura mujeres'")
plt.title("Diagrama de dispersión de peso mujeres vs.cintura mujeres")

plt.show()
```
