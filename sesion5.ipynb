```python
df=pd.read_csv('datasets/Cartwheeldata.csv')
df.head(2)
```

```python
df.columns
```

```python
import seaborn as sns

sns.lmplot(
    x='Wingspan',
    y='CWDistance',
    data=df,
    fit_reg=False,
    hue='Gender'
)
plt.show()
```

```python
sns.swarmplot(
    x='Gender',
    y='CWDistance',
    data = df
)
plt.show()
```

```python
sns.boxplot(
    data=df.loc[:,['Height','Wingspan','CWDistance']]
)
plt.show()
```

```python
fig=plt.figure()

ax1=fig.add_subplot(1,2,1)
sns.boxplot(
    data=df.loc[df['Gender']=='M',['Score']]
)

ax2=fig.add_subplot(1,2,2)
sns.boxplot(
    data=df.loc[df['Gender']=='F',['Score']]
)

ax1.title.set_text('Male boxplot')
ax2.title.set_text('Female boxplot')

plt.show()
```
